```python
from matplotlib import pyplot as plt
import cv2
import numpy as np

num_image=165
im_origin = []
new_im = []

file = '/Users/macbookretina/Desktop/Github_Images_Biomdicales/projet_images_biomedicales/Banque_images_segmentation/p0/Itraining_'+str(num_image)+'.bmp'
im_origin.append(cv2.imread(file))
gray = cv2.cvtColor(im_origin[0], cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray, 65, 255, cv2.THRESH_BINARY)
kernel = np.ones((3,3),np.uint8)
kernel2 = np.ones((2,2),np.uint8)
closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel,iterations = 1)
erosion = cv2.erode(closing,kernel2,iterations = 2)
opening = cv2.morphologyEx(erosion, cv2.MORPH_OPEN, kernel,iterations = 1)
erosion2 = cv2.erode(opening,kernel2,iterations = 2)
    
file_to_compare = '/Users/macbookretina/Desktop/Github_Images_Biomdicales/projet_images_biomedicales/Images Cardiaques-20201208/ShortAxisSlices/Patient0/Igroundtruth_'+str(num_image)+'.bmp'
new_im.append(cv2.imread(file_to_compare))

fig = plt.figure()    
ax = fig.add_subplot(1, 3, 1)
imgplot = plt.imshow(im_origin[0])
ax.set_title('Itraining_'+str(num_image)+'.bmp')
ax = fig.add_subplot(1, 3, 2)
imgplot = plt.imshow(erosion2)
ax.set_title('Image segmentee')
ax = fig.add_subplot(1, 3, 3)
imgplot = plt.imshow(new_im[0])
ax.set_title('Igroundtruth_'+str(num_image)+'.bmp')
    
num_image = num_image + 1 
    
# Filename 
filename = 'savedImage' + str(num_image) + '.bmp'
# Using cv2.imwrite() method str
# Saving the image 
cv2.imwrite(filename, erosion2) 
    
```
